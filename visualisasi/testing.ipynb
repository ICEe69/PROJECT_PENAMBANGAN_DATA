```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Ganti dengan nama file Anda
file_name = 'nilai82.csv'  # Sesuaikan nama file Anda

# Membaca data dari file CSV
df = pd.read_csv(file_name)

# Pastikan data yang dibaca adalah numerik (jika ada kolom yang tidak numerik, konversi ke tipe numerik)
df['UTS'] = pd.to_numeric(df['UTS'], errors='coerce')
df['SAS'] = pd.to_numeric(df['SAS'], errors='coerce')

# Menghapus baris dengan nilai NaN
df = df.dropna()
```

```python
# Tentukan jumlah klaster yang diinginkan (misalnya 3 klaster)
num_clusters = 3

# Membuat model K-Means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Fitting model K-Means ke data
kmeans.fit(df[['UTS', 'SAS']])

# Menyimpan hasil klastering ke kolom baru 'Cluster'
df['Cluster'] = kmeans.labels_

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
plt.scatter(df['UTS'], df['SAS'], c=df['Cluster'], cmap='viridis', marker='o', alpha=0.7)
plt.xlabel('UTS')
plt.ylabel('SAS')
plt.title('Clustering dengan K-Means')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Menentukan inertia untuk berbagai jumlah klaster
inertia = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df[['UTS', 'SAS']])
    inertia.append(kmeans.inertia_)

# Plot Elbow Method untuk mencari jumlah klaster optimal
plt.figure(figsize=(8, 5))
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Jumlah Klaster')
plt.ylabel('Inertia')
plt.title('Elbow Method untuk Menentukan Jumlah Klaster Optimal')
plt.show()
```
